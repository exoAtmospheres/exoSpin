## AB Pic b Plot - Sid papers


#### **1 - Imports**


```python
# Imports
import exoSpin as xs
```

#### **2 - Inputs**

Plot for $P_{rot} = [2.1 , 4.2, 8.4, 16.8, 33.6]$

```python
# File paths & values for AB Pic b.

n = 10000 # Number of samples wanted to generate a normal distribution from only mean value and standard deviation. 

exoplanet_name = 'AB Pic b'
io = 'data/io_2024.dat'
radius = 'data/rp_vrot.txt'
vsini = 'data/rp_vrot.txt'
omega = [-5,4,n]

P_1 = 2.1
P_2 = 4.2
P_3 = 8.4
P_4 = 16.8
P_5 = 33.6

M = 10
```

#### **3 - Obliquity()**

```python
x_1 = xs.obliquity(exoplanet_name,io,radius,vsini,omega,P_1,M)
#x_2 = xs.obliquity(exoplanet_name,io,radius,vsini,omega,P_2,M)
x_3 = xs.obliquity(exoplanet_name,io,radius,vsini,omega,P_3,M)
x_4 = xs.obliquity(exoplanet_name,io,radius,vsini,omega,P_4,M)
#x_5 = xs.obliquity(exoplanet_name,io,radius,vsini,omega,P_5,M)
```

```python
from matplotlib.lines import Line2D
import numpy as np
from scipy.stats import gaussian_kde

# Setting plot for i_p with P_rot = [2.1 , 4.2, 8.4, 16.8, 33.6]
ip_1_pdf = x_1.ip_pdf_saved
x_ip_1 = ip_1_pdf.x
y_ip_1 = ip_1_pdf.y
mean_ip_1 = np.trapz(x_ip_1*y_ip_1, x_ip_1)
var_ip_1 = np.sqrt(np.trapz((x_ip_1-mean_ip_1)**2 * y_ip_1, x_ip_1))

# ip_2_pdf = x_2.ip_pdf_saved
# x_ip_2 = ip_2_pdf.x
# y_ip_2 = ip_2_pdf.y
# mean_ip_2 = np.trapz(x_ip_2*y_ip_2, x_ip_2)
# var_ip_2 = np.sqrt(np.trapz((x_ip_2-mean_ip_2)**2 * y_ip_2, x_ip_2))

ip_3_pdf = x_3.ip_pdf_saved
x_ip_3 = ip_3_pdf.x
y_ip_3 = ip_3_pdf.y
mean_ip_3 = np.trapz(x_ip_3*y_ip_3, x_ip_3)
var_ip_3 = np.sqrt(np.trapz((x_ip_3-mean_ip_3)**2 * y_ip_3, x_ip_3))

ip_4_pdf = x_4.ip_pdf_saved
x_ip_4 = ip_4_pdf.x
y_ip_4 = ip_4_pdf.y
mean_ip_4 = np.trapz(x_ip_4*y_ip_4, x_ip_4)
var_ip_4 = np.sqrt(np.trapz((x_ip_4-mean_ip_4)**2 * y_ip_4, x_ip_4))

# ip_5_pdf = x_5.ip_pdf_saved
# x_ip_5 = ip_5_pdf.x
# y_ip_5 = ip_5_pdf.y
# mean_ip_5 = np.trapz(x_ip_5*y_ip_5, x_ip_5)
# var_ip_5 = np.sqrt(np.trapz((x_ip_5-mean_ip_5)**2 * y_ip_5, x_ip_5))

# Setting plot for i_o
io_pdf = x_1.pdf('Orbital inclination')
mean_io = np.trapz(io_pdf.x * io_pdf.y, io_pdf.x)
var_io = np.sqrt(np.trapz((io_pdf.x-mean_io)**2 * io_pdf.y, io_pdf.x))

io_file = open('data/io.dat', "r")
io_samp_old = np.loadtxt(io_file, skiprows=1)
io_samp_old = np.rad2deg(io_samp_old)
angles = np.linspace(0,180,1000)
io_old_kde = gaussian_kde(io_samp_old,bw_method='scott')
io_old_pdf = io_old_kde(angles)
mean_io_old = np.trapz(angles * io_old_pdf, angles)
var_io_old = np.sqrt(np.trapz((io_old_pdf-mean_io_old)**2 * io_old_pdf, angles))


# Setting plot for Psi_proj
pro_1_pdf = x_1.pdf('Projected obliquity - complex')
mean_pro_1 = np.trapz(pro_1_pdf.x * pro_1_pdf.y, pro_1_pdf.x)
var_pro_1 = np.sqrt(np.trapz((pro_1_pdf.x-mean_pro_1)**2 * pro_1_pdf.y, pro_1_pdf.x))

# pro_2_pdf = x_2.pdf('Projected obliquity - complex')
# mean_pro_2 = np.trapz(pro_2_pdf.x * pro_2_pdf.y, pro_2_pdf.x)
# var_pro_2 = np.sqrt(np.trapz((pro_2_pdf.x-mean_pro_2)**2 * pro_2_pdf.y, pro_2_pdf.x))

pro_3_pdf = x_3.pdf('Projected obliquity - complex')
mean_pro_3 = np.trapz(pro_3_pdf.x * pro_3_pdf.y, pro_3_pdf.x)
var_pro_3 = np.sqrt(np.trapz((pro_3_pdf.x-mean_pro_3)**2 * pro_3_pdf.y, pro_3_pdf.x))

pro_4_pdf = x_4.pdf('Projected obliquity - complex')
mean_pro_4 = np.trapz(pro_4_pdf.x * pro_4_pdf.y, pro_4_pdf.x)
var_pro_4 = np.sqrt(np.trapz((pro_4_pdf.x-mean_pro_4)**2 * pro_4_pdf.y, pro_4_pdf.x))

# pro_5_pdf = x_5.pdf('Projected obliquity - complex')
# mean_pro_5 = np.trapz(pro_5_pdf.x * pro_5_pdf.y, pro_5_pdf.x)
# var_pro_5 = np.sqrt(np.trapz((pro_5_pdf.x-mean_pro_5)**2 * pro_5_pdf.y, pro_5_pdf.x))




```

```python
import matplotlib.pyplot as plt

#### Legends for every parameters
plt.rcParams.update({"font.size": 16})
io_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='lightcoral',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

io_old_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

ip_1_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='yellow',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.7)
                
ip_2_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='gold',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.7)

ip_3_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='goldenrod',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.7)

ip_4_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='darkgoldenrod',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.5)

ip_5_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='olive',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.7)

Psi_1_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

Psi_2_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

Psi_3_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

Psi_4_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

Psi_5_legend = Line2D([0], [0], color='black', linestyle='None',
                        marker='s', markersize=15, markerfacecolor='grey',
                        markeredgewidth=1.5, markeredgecolor='black',alpha=0.3)

#### Plots setting
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,4))

#### Plot of i_o
ax1.plot(angles,io_old_pdf, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax1.fill_between(angles, io_old_pdf, color='grey', alpha=0.3)
ax1.plot(io_pdf.x,io_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax1.fill_between(io_pdf.x,io_pdf.y, color='lightcoral', alpha=0.3)
for spine in ax1.spines.values():
    spine.set_linewidth(3)
ax1.tick_params(axis='x', which='both', direction='in', length=10, width=1, top=True)
ax1.tick_params(axis='y', which='both', direction='in', length=0, width=1)
ax1.set_yticklabels([])
ax1.set_ylim(bottom=0)
ax1.set_xlabel('$i_o$/deg', fontsize=16)
ax1.errorbar(mean_io, io_pdf.y[569]*0.35, xerr=var_io, yerr=0, fmt='^', color='coral', capsize=5)
ax1.errorbar(mean_io_old, io_old_pdf[500]*0.16, xerr=var_io, yerr=0, fmt='D', color='grey', capsize=5)
ax1.set_ylabel('Probability Density')
ax1.legend([io_legend, io_old_legend],
           ['Updated posteriors', 'P.Palma-Bifani et al. 2023'], loc='upper left', frameon=False,fontsize=14)

# Plot of i_p
ax2.plot(ip_1_pdf.x,ip_1_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax2.fill_between(ip_1_pdf.x,ip_1_pdf.y, color='yellow', alpha=0.7)
# ax2.plot(ip_2_pdf.x,ip_2_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
# ax2.fill_between(ip_2_pdf.x,ip_2_pdf.y, color='gold', alpha=0.7)
ax2.plot(ip_3_pdf.x,ip_3_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax2.fill_between(ip_3_pdf.x,ip_3_pdf.y, color='goldenrod', alpha=0.7)
ax2.plot(ip_4_pdf.x,ip_4_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax2.fill_between(ip_4_pdf.x,ip_4_pdf.y, color='darkgoldenrod', alpha=0.5)
# ax2.plot(ip_5_pdf.x,ip_5_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
# ax2.fill_between(ip_5_pdf.x,ip_5_pdf.y, color='olive', alpha=0.7)
ax2.legend([ip_1_legend, ip_3_legend, ip_4_legend ],
           ['$P_{rot}= 2.1 hr$', '$P_{rot}= 8.4 hr$', '$P_{rot}= 16.8 hr$'], loc='upper center', frameon=False,fontsize=14)

for spine in ax2.spines.values():
    spine.set_linewidth(3)
ax2.tick_params(axis='x', which='both', direction='in', length=10, width=1, top=True)
ax2.tick_params(axis='y', which='both', direction='in', length=0, width=1)
ax2.set_xlabel('$i_p$/deg', fontsize=16)

ax2.set_yticklabels([])
ax2.set_ylim(bottom=0)
#ax2.errorbar(mean_ip, ip_pdf.y[2]*0.2, xerr=var_ip, yerr=0, fmt='o', color='gold', capsize=5)

# Plot of Psi_proj
ax3.plot(pro_1_pdf.x,pro_1_pdf.y, color='black', alpha=0.5, linewidth=1,linestyle="dashdot")
ax3.fill_between(pro_1_pdf.x,pro_1_pdf.y, color='cyan', alpha=0.3)
for spine in ax3.spines.values():
    spine.set_linewidth(3)
ax3.tick_params(axis='x', which='both', direction='in', length=10, width=1, top=True)
ax3.tick_params(axis='y', which='both', direction='in', length=0, width=1)
ax3.set_yticklabels([])
ax3.set_ylim(bottom=0)
ax3.set_xlabel('$\Psi_{projected}=|i_p - i_o|$/deg', fontsize=16)
ax3.errorbar(mean_pro_1, pro_1_pdf.y[48]*0.15, xerr=var_pro_1, yerr=0, fmt='s', color='c', capsize=5)

plt.show()
fig.savefig('ab_pic_paper_plot.pdf', format='pdf',bbox_inches='tight')


```
